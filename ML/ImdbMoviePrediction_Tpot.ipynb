```python
from tpot import TPOTRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
```

```python
#read data
data = pd.read_csv('./movie_metadata.csv',header='infer',usecols=["gross","num_voted_users","budget", "imdb_score"])
data.columns = ["class","votes","budget","rating"]
df = pd.DataFrame(data)
df_target = df[["class"]]
df_data = df[["votes","budget","rating"]]
```

```python
# scaler = MinMaxScaler(feature_range=(0, 1))
# scaled_values = scaler.fit_transform(df_data.values)
# df_data_scaled = pd.DataFrame(scaled_values, columns=df_data.columns)
# df_data_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_data, df_target,
                                                    train_size=0.75, test_size=0.25,random_state=1234)
```

```python
X_train.fillna(X_train.mean(), inplace=True)
X_test.fillna(X_test.mean(), inplace=True)
y_train.fillna(y_train.mean(),inplace=True)
y_test.fillna(y_train.mean(),inplace=True)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
tpot = TPOTRegressor(generations=5, population_size=20, verbosity=2)
tpot.fit(X_train, y_train)
print(tpot.score(X_test, y_test))
predicted = tpot.predict(X_test)
tpot.export('tpot_boston_pipeline.py')
```

```python
result = pd.DataFrame({"expected_value":y_test["class"].to_numpy(),"predicted_value":predicted})
result
```

```python
result.to_csv('./results.csv')
```

```python
mse_sum = 0
for index, row in result.iterrows():
    mse_sum += (row["expected_value"] - row["predicted_value"])**2
mse = mse_sum / len(result)
print(mse)
```
